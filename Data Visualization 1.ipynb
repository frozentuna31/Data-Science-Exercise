### Answer :

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn import datasets
from sklearn.decomposition import PCA
```

```python
#Preparing the data
iris= datasets.load_iris()
df = pd.DataFrame(data= iris['data'],
                  columns= iris['feature_names'])
spec = pd.DataFrame(data = iris.target, columns = ['Species'])
df
```

```python
#Visualization
pca = PCA(n_components=3)
pca.fit(df)
df_pca = pca.transform(df)
fig1 = plt.figure(figsize = (7,7))
ax1 = Axes3D(fig1, rect=[1,1,1,1], elev = 35, azim = 75)
ax1.scatter(df_pca[:,0], df_pca[:,1], df_pca[:,2], c = spec, cmap = 'plasma',edgecolors = 'black')

print('Yellow is Setosa \nRed is Versicolour \nBlue is Virginica')
```
